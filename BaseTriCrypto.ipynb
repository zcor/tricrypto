```python
import pandas as pd

# Load the data
file_path = 'tricrypto_base-2024-01-09_17-01-58.csv'
data = pd.read_csv(file_path)
data
```

```python
# Plotting all columns against each other
import matplotlib.pyplot as plt

pd.plotting.scatter_matrix(data, alpha=0.2, figsize=(10, 10))
plt.show()
data
```

```python
# Creating a plot with "Block" against " A" and adding " Fee" on a new axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting "Block" vs " A" as a dashed line
color = 'tab:blue'
ax1.set_xlabel('Block')
ax1.set_ylabel('Fee Gamma', color=color)
ax1.plot(data['Block'], data['Fee_Gamma'], color=color, linestyle='dashed', alpha=0.5)  # Changed to plot with dashed line
ax1.tick_params(axis='y', labelcolor=color)

# Creating a new axis for " Fee"
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Fee', color=color)
ax2.plot(data['Block'], data['Fee'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # to ensure the right y-label is not slightly clipped
plt.title('Effect of Parameter Changes on Base TriCrypto Fees')
plt.grid(True)
plt.show()
```
